### âœ… 1. Basic List Comprehension Syntax

List comprehension is a short, clean, and faster way to create lists in Python.

```python
## âœ” General Syntax:
new_list = [expression for item in iterable]
```

### Meaning:

#### expression â†’ what you want to store in the new list

### item â†’ each element from the iterable

### iterable â†’ something you loop through (list, range, string, etc.)

### without List comprehention

```python
numbers = [1, 2, 3, 4, 5]
squares = []

for n in numbers:
    squares.append(n * n)

print(squares)   # [1, 4, 9, 16, 25]
```

### with list comprehension

```python
numbers = [1, 2, 3, 4, 5]
squares = [n * n for n in numbers]

print(squares)   # [1, 4, 9, 16, 25]
```

### âœ” Example 2: Filter even numbers

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = [n for n in numbers if n % 2 == 0]

print(evens)   # [2, 4, 6]
```

### Even and odd

```python
numbers = [1, 2, 3, 4, 5, 6]

result = [(n, "Even") if n % 2 == 0 else (n, "Odd") for n in numbers]

print(result)
# Output: [(1, 'Odd'), (2, 'Even'), (3, 'Odd'), (4, 'Even'), (5, 'Odd'), (6, 'Even')]
```

### Extract first letter of each word

```python
words = ["dog", "cat", "lion"]
first_letters = [word[0] for word in words]
print(first_letters)   # ['d', 'c', 'l']
```
