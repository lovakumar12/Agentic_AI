#### 1. What is a Module?

A module in Python is simply a file that contains Python code (functions, variables, classes).
Modules help us:

Reuse code

Organize large programs

Use built-in functionality without writing everything ourselves

#### ✅ 2. Importing Modules
Syntax 1: import module_name

This imports the entire module.

```python
import math

result = math.sqrt(25)
print(result)
```

#### ✅ 3. from…import Syntax
Syntax 2: from module_name import specific_item

This allows us to import only what we need.

```python
from math import sqrt

print(sqrt(36))
```

#### 4. Creating Simple Modules

#### create using .py file

```python
# mymath.py
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b
```

```python
import mymath

print(mymath.add(10, 20))
print(mymath.multiply(5, 6))
```

```python
import math

print(math.pi)
print(math.floor(4.9))
print(math.ceil(4.1))
```

```python
import random

print(random.randint(1, 10))   # Random number 1–10
colors = ["red", "blue", "green"]
print(random.choice(colors))   # Random color
```

#### 7. datetime Basics

Used to work with dates and time.

```python
from datetime import datetime

now = datetime.now()
print(now)
```

```python
from datetime import datetime

now = datetime.now()
print(now.year)
print(now.month)
print(now.day)
```
