## 1. Installing Python and Package Managers

**Python installation (Windows / macOS / Linux)**

- Recommended: download from the official website (https://www.python.org/downloads/).
- On Windows, **check** "Add Python to PATH" during installation.
- Verify installation in a terminal / command prompt:
```
python --version
pip --version
```

**Package managers**

- `pip` is installed with Python and is used to install extra libraries:
```
pip install requests
```
- Optional: `conda` (from Anaconda / Miniconda) provides Python + many scientific packages together.

Students only need basic comfort with installing Python and using `pip install <package>` from a terminal.

## 2. Virtual Environments Basics

Virtual environments help isolate dependencies for each project.

### Creating and activating a virtual environment (built‚Äëin `venv`)

# Create env named .venv
python -m venv .venv

# Activate (Windows PowerShell)
.venv\\Scripts\\Activate

# Activate (macOS / Linux)
source .venv/bin/activate

# Install packages inside env
pip install numpy


- After activation, `python` and `pip` commands use the environment.
- Deactivate with the `deactivate` command.
